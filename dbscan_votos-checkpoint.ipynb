```python
from sklearn.preprocessing import normalize
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score\
```

```python
del votos['Candidato']
```

```python
X = votos.as_matrix()
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(
    eps = 0.8, # variar
    min_samples = 19 # variar
)
print(dbscan)
```

```python
dbscan.fit(X)
```

```python
outliers = pd.DataFrame(votos)
```

```python
print(Counter(dbscan.labels_))

```

```python
print(outliers[dbscan.labels_==-1])
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
colors = dbscan.labels_
ax.scatter(votos['Despesas'], votos['Votos'], c=colors,  s=120)
ax.set_xlabel('Despesa')
ax.set_ylabel('Votos')
plt.title('Detecção anomalias')
```
