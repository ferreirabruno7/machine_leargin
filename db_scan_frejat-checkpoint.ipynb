```python
from sklearn.preprocessing import normalize
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
X = frejat.as_matrix()
```

```python
X
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(
    eps = 0.5, # variar
    min_samples = 5 # variar
)
print(dbscan)
```

```python
dbscan.fit(X)
```

```python
outliers = pd.DataFrame(frejat)
```

```python
print(Counter(dbscan.labels_))

```

```python
print(outliers[dbscan.labels_==-1])
```

```python
frejat['cluster'] = dbscan.labels_ # adiciona coluna c/ cluster
```

```python
silhouette_avg = silhouette_score(X, dbscan.labels_)
```

```python
print(silhouette_avg)
```

```python
for k in set(dbscan.labels_):
    print(frejat[frejat['cluster'] == k])
```
