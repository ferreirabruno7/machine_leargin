```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
votos = pd.read_csv("C:/Users/NB/Desktop/eleitorado/base_votos_2014.csv", delimiter = ',', encoding='utf-8', decimal=',' )
```

```python
votos['Despesas'].fillna(value=votos['Despesas'].median(), inplace=True)
votos['Valor Corrigido'].fillna(value=votos['Valor Corrigido'].median(), inplace=True)
votos['Custo do Voto'].fillna(value=votos['Custo do Voto'].median(), inplace=True)
votos['Custo Corrigido'].fillna(value=votos['Custo Corrigido'].median(), inplace=True)
```

```python
votos
```

```python
X = votos[[
    'Valor Corrigido',
    'Votos'
]]
```

```python
X = normalize(X.as_matrix())
```

```python
kmeans = KMeans(
    n_clusters = 3, # testar diferentes valores
    n_init = 15, 
    max_iter = 450, 
    init = 'k-means++'
)
```

```python
kmeans.fit(X)
```

```python
votos['cluster'] = kmeans.labels_ # adiciona coluna c/ cluster
grouped = votos.groupby('cluster')
stacked = pd.DataFrame()
for group in grouped:
    means = group[1].mean(axis = 0)
    stacked[group[0]] = means
print(stacked)
```

```python
silhouette_avg = silhouette_score(X, kmeans.labels_)
print('silhueta media:', silhouette_avg)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(X[:, 0], X[:,1], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'red',label = 'Centroids')
plt.title('Clusterização Votos x Despesa')
plt.xlabel('Despesa')
plt.ylabel('Voto')
plt.legend()

plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = votos[['Despesas', 'Valor Corrigido', 'Custo do Voto',
       'Custo Corrigido']]
```

```python
y = votos['Coligação']
```

```python
X_train, X_test, y_tran, y_test = train_test_split(X,y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression()
```

```python
lm.fit(X_train, y_tran)
```

```python
predict = lm.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predict))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, predict))
```
