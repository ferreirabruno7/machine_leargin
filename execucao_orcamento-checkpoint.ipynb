```python
x = orc[['Empenhado']]
```

```python
y = orc['Total Pago']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_tran, y_test = train_test_split(x,y, test_size=0.3, random_state=101)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_tran)

```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
coefs = pd.DataFrame(lm.coef_, x.columns, columns=['coefs'])
```

```python
coefs
```

```python
predict = lm.predict(X_test)
```

```python
#aderencia entre o predito e treino, eixo x é o valor efetivo de votos e eixo Y o resultado dos testes
plt.figure(figsize=(12,6))
plt.xlabel('Valores Reais')
plt.ylabel('Valores Preditos')
plt.scatter(y_test, predict)
```

```python
#distribuição normal do votos, o modelo tem um grande desvio padrão
sns.distplot((y_test-predict))
```

```python
from sklearn import metrics
```

```python
##média de erro do modelo na predição de votos, quanto o modelo erra, ele erra em 1191 votos.
print('MAE', metrics.mean_absolute_error(y_test, predict))
```

```python
##média de erro do modelo na predição de votos, quanto o modelo erra, ele erra em 1191 votos.
print('MSE', metrics.mean_squared_error(y_test, predict))
```

```python
import numpy as np
import pandas as pd
import sys
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
%matplotlib inline
```

```python
X = orc[[
    
    'Empenhado',    
'Liquidado',     
'Pago EX',       
'Pago RPP',      
'Pago RPNP',     
'Pago RET',      
'Total Pago' 
]]
kmeans = KMeans(
    n_clusters = 7, # testar diferentes valores
    n_init = 15, 
    max_iter = 450, 
    init = 'k-means++'
)
kmeans.fit(X)
```

```python
orc
```

```python

orc['cluster'] = kmeans.labels_ # adiciona coluna c/ cluster
grouped = orc.groupby('cluster')
stacked = pd.DataFrame()
for group in grouped:
    means = group[1].mean(axis = 0)
    
    stacked[group[0]] = means
print(stacked)
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
# k means determine k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
 
# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distorção ')
plt.title('Método Elbow Frejat otimização K')
plt.show()
 
```
