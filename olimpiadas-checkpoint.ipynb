```python
import pandas as pd
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
##importação do arquivo CSV
olimp = pd.read_csv("C:/Users/NB/Desktop/desafio/summer.csv", delimiter = ',', encoding='iso-8859-1', error_bad_lines=False, decimal='.', usecols = [
        'Sport',
        'Country',
    ]  )
```

```python
olimp
```

```python
olimp['1'] = 1
```

```python
olimp
```

```python
olimp = olimp.pivot_table(
    index = 'Country', 
    columns = 'Sport', 
    aggfunc = 'sum',
    fill_value = 0
)
```

```python
olimp
```

```python
# "achata" DataFrame multi-indice
cols = [e[1] for e in olimp.columns]
rows = list(olimp.index)
olimp = pd.DataFrame(olimp.as_matrix())
olimp.columns = cols
```

```python
olimp
```

```python
X = olimp.as_matrix()
```

```python
num_clusters = 3
kmeans = KMeans(
    n_clusters = num_clusters, # testar diferentes valores
    n_init = 150, 
    max_iter = 450, 
    init = 'k-means++'
)
```

```python
kmeans.fit(X)
```

```python
olimp['cluster'] = kmeans.labels_
olimp['country'] = rows
for k in range(num_clusters):
    print(olimp[olimp['cluster'] == k])
silhouette_avg = silhouette_score(X, kmeans.labels_)
print('silhueta media:', silhouette_avg)
```
