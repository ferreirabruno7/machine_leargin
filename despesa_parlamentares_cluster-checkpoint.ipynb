```python
import pandas as pd
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
##importação do arquivo CSV
olimp = pd.read_csv("C:/Users/NB/Desktop/BCK/despesas_parlamentares/Ano-2018.csv", delimiter = ';', encoding='utf-8', error_bad_lines=False, decimal='.', usecols = [
        'sgUF',
        'sgPartido',
        'vlrLiquido'
    ])
```

```python
olimp.head().T
```

```python
olimp.info()
```

```python
olimp['vlrLiquido'] = pd.to_numeric(olimp['vlrLiquido'], errors='coerce')

```

```python
olimp['vlrLiquido'].fillna(value=olimp['vlrLiquido'].median(), inplace=True)
olimp
```

```python
olimp = olimp.pivot_table(
    index = 'sgUF', 
    columns = 'sgPartido', 
    aggfunc = 'sum',
    fill_value = 0
)
```

```python
olimp
```

```python
# "achata" DataFrame multi-indice
cols = [e[1] for e in olimp.columns]
rows = list(olimp.index)
olimp = pd.DataFrame(olimp.as_matrix())
olimp.columns = cols
```

```python
olimp
```

```python
X = olimp.as_matrix()
```

```python
num_clusters = 4
kmeans = KMeans(
    n_clusters = num_clusters, # testar diferentes valores
    n_init = 150, 
    max_iter = 450, 
    init = 'k-means++'
)
```

```python
kmeans.fit(X)
```

```python
olimp['cluster'] = kmeans.labels_
olimp['sgUF'] = rows
for k in range(num_clusters):
    print(olimp[olimp['cluster'] == k])
silhouette_avg = silhouette_score(X, kmeans.labels_)
print('silhueta media:', silhouette_avg)
```

```python
## importação das bibliotecas de para clusterização
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
##Importação das bibliotecas
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 

#importação bibliotecas metodo elbow
from scipy.spatial.distance import cdist

# k means determine k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
 
# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distorção ')
plt.title('Método Elbow Ambev otimização K')
plt.show()
 
```
