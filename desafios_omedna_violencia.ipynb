```
desafios['1'] = 1
```

```
desafios = desafios.pivot_table(
    index = 'Country', 
    columns = 'CATEGORY', 
    aggfunc = 'sum',
    fill_value = 0
)
```

```
desafios.head(5)
```

```
# "achata" DataFrame multi-indice
cols = [e[1] for e in desafios.columns]
rows = list(desafios.index)
desafios = pd.DataFrame(desafios.as_matrix())
desafios.columns = cols
```

```
desafios
```

```
# prepara dados
X = desafios.as_matrix()
```

```
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```
# inicializa clusterizador
num_clusters = 3
kmeans = KMeans(
    n_clusters = num_clusters, # testar diferentes valores
    n_init = 150, 
    max_iter = 450, 
    init = 'k-means++'
)
```

```
# clusteriza
kmeans.fit(X)
```

```
# checa silhuetas
desafios['cluster'] = kmeans.labels_
desafios['country'] = rows
for k in range(num_clusters):
    print(desafios[desafios['cluster'] == k])
silhouette_avg = silhouette_score(X, kmeans.labels_)
print('silhueta media:', silhouette_avg)
```
