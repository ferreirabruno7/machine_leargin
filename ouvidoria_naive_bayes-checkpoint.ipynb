```python
from sklearn.model_selection import train_test_split
```

```python
X = ouvidoria[[
    
    
    'Assunto_id',
    'IDADE',
    'Sexo_id',
    'Classificacao_id', 
    'RegiaoAdministrativa_id'
   
]]

y = ouvidoria['Situacao_id']
```

```python
X_train, X_test, y_tran, y_test = train_test_split(X,y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_tran)
```

```python
pred = dtree.predict(X_test)
```

```python
print(dtree)
y_expect = y_test
previsores2 = dtree.predict(X_test)
precisao2 = accuracy_score(y_test, pred)
print(precisao2)
print(classification_report(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rcf = RandomForestClassifier(n_estimators=500)
```

```python
rcf.fit(X_train, y_tran)
```

```python
pred = rcf.predict(X_test)
```

```python
print(rcf)
y_expect = y_test
previsores2 = rcf.predict(X_test)
precisao2 = accuracy_score(y_test, pred)
print(precisao2)
print(classification_report(y_test, pred))
```

```python
resultado = rcf.predict([[525,42,0, 3,2]])
print(rcf.classes_)
print(resultado)
```
